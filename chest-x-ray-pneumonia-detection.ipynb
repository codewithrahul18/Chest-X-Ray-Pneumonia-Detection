```python


# ================================================================
# 1. Imports
# ================================================================

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from tqdm import tqdm

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam

# ================================================================
# 2. Dataset Path Setup
# ================================================================

BASE_DIR = "/kaggle/input/chest-xray-pneumonia/chest_xray"

TRAIN_DIR = os.path.join(BASE_DIR, "train")
TEST_DIR = os.path.join(BASE_DIR, "test")
VAL_DIR = os.path.join(BASE_DIR, "val")

classes = ["NORMAL", "PNEUMONIA"]

# ================================================================
# 3. Dataset Exploration
# ================================================================

def count_images(directory):
    data = {}
    for cls in classes:
        path = os.path.join(directory, cls)
        data[cls] = len(os.listdir(path))
    return data

train_counts = count_images(TRAIN_DIR)
test_counts = count_images(TEST_DIR)
val_counts = count_images(VAL_DIR)

counts_df = pd.DataFrame({
    "Train": train_counts,
    "Validation": val_counts,
    "Test": test_counts
})

print(counts_df)

# ================================================================
# 4. Dataset Distribution Visualization
# ================================================================

counts_df.T.plot(kind="bar", figsize=(10,6))
plt.title("Class Distribution Across Dataset")
plt.ylabel("Number of Images")
plt.grid(alpha=0.3)
plt.show()

# ================================================================
# 5. Sample Image Visualization
# ================================================================

plt.figure(figsize=(10,6))

for i, cls in enumerate(classes):
    img_path = os.path.join(TRAIN_DIR, cls, os.listdir(os.path.join(TRAIN_DIR, cls))[0])
    img = Image.open(img_path)

    plt.subplot(1,2,i+1)
    plt.imshow(img, cmap="gray")
    plt.title(cls)
    plt.axis("off")

plt.show()

# ================================================================
# 6. Image Preprocessing & Augmentation
# ================================================================

IMG_SIZE = 224
BATCH_SIZE = 32

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,
    zoom_range=0.2,
    horizontal_flip=True
)

val_test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    TRAIN_DIR,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary'
)

val_generator = val_test_datagen.flow_from_directory(
    VAL_DIR,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary'
)

test_generator = val_test_datagen.flow_from_directory(
    TEST_DIR,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=False
)

# ================================================================
# 7. CNN Model Architecture
# ================================================================

model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(IMG_SIZE, IMG_SIZE, 3)),
    BatchNormalization(),
    MaxPooling2D(2,2),

    Conv2D(64, (3,3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D(2,2),

    Conv2D(128, (3,3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D(2,2),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

model.compile(
    optimizer=Adam(learning_rate=0.0001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()

# ================================================================
# 8. Model Training
# ================================================================

EPOCHS = 10

history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=EPOCHS
)

# ================================================================
# 9. Training Curves Visualization
# ================================================================

plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.title('Accuracy Curve')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Loss Curve')
plt.legend()

plt.show()

# ================================================================
# 10. Model Evaluation
# ================================================================

pred_probs = model.predict(test_generator)
pred_labels = (pred_probs > 0.5).astype(int).ravel()
true_labels = test_generator.classes

print(classification_report(true_labels, pred_labels, target_names=classes))

# ================================================================
#  Model Performance Analysis (Class Imbalance Issue)
# ================================================================

"""
Your classification report shows:

NORMAL    recall = 0.09    (very poor)
PNEUMONIA recall = 1.00    (very high)

This means:
- The model predicts almost everything as **PNEUMONIA**
- It is missing most NORMAL cases

This happens because the dataset is **highly imbalanced**:
PNEUMONIA images >> NORMAL images

In medical AI, this is dangerous because healthy patients are misclassified.

Below are professional fixes used in real projects.
"""

# ================================================================
#  Solution 1  Use Class Weights (Most Important Fix)
# ================================================================

from sklearn.utils.class_weight import compute_class_weight

# Get class labels from training generator
labels = train_generator.classes

class_weights_array = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(labels),
    y=labels
)

class_weights = {0: class_weights_array[0], 1: class_weights_array[1]}

print("Class Weights:", class_weights)

# Re-train model using class weights

history_weighted = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=10,
    class_weight=class_weights
)

# ================================================================
#  Solution 2  Adjust Decision Threshold
# ================================================================

"""
Default threshold = 0.5
We increase it to 0.7 to reduce false pneumonia predictions.
"""

new_threshold = 0.7

pred_labels_adjusted = (pred_probs > new_threshold).astype(int).ravel()

print("Classification Report (Adjusted Threshold):")
print(classification_report(true_labels, pred_labels_adjusted, target_names=classes))

# ================================================================
#  Solution 3  Evaluate Improved Confusion Matrix
# ================================================================

cm_adjusted = confusion_matrix(true_labels, pred_labels_adjusted)

plt.figure(figsize=(6,5))
sns.heatmap(cm_adjusted, annot=True, fmt='d', cmap='Greens',
            xticklabels=classes, yticklabels=classes)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix (After Fixes)")
plt.show()

# ================================================================
# 11. Confusion Matrix
# ================================================================

cm = confusion_matrix(true_labels, pred_labels)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

# ================================================================
# 12. Prediction Visualization
# ================================================================

sample_indices = np.random.choice(len(test_generator.filenames), 6, replace=False)

plt.figure(figsize=(12,6))

for i, idx in enumerate(sample_indices):
    img_path = os.path.join(TEST_DIR, test_generator.filenames[idx])
    img = Image.open(img_path).resize((IMG_SIZE, IMG_SIZE))

    pred = "PNEUMONIA" if pred_labels[idx] == 1 else "NORMAL"
    true = classes[true_labels[idx]]

    plt.subplot(2,3,i+1)
    plt.imshow(img)
    plt.title(f"True: {true}\nPred: {pred}")
    plt.axis('off')

plt.tight_layout()
plt.show()

# ================================================================
# 13. Save Model
# ================================================================

model.save("pneumonia_detector_model.h5")

print("Model saved successfully!")

# ================================================================
# 14. Advanced Model  Transfer Learning (ResNet50)
# ================================================================

from tensorflow.keras.applications import ResNet50
from tensorflow.keras.models import Model
from tensorflow.keras.layers import GlobalAveragePooling2D

base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3))

for layer in base_model.layers:
    layer.trainable = False

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.5)(x)
output = Dense(1, activation='sigmoid')(x)

transfer_model = Model(inputs=base_model.input, outputs=output)

transfer_model.compile(
    optimizer=Adam(learning_rate=0.0001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

transfer_model.summary()

# ================================================================
# 15. Train Transfer Learning Model
# ================================================================

history_tl = transfer_model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=5
)

# ================================================================
# 16. Grad-CAM Explainability
# ================================================================

import cv2

def make_gradcam_heatmap(img_array, model, last_conv_layer_name="conv5_block3_out"):
    grad_model = tf.keras.models.Model(
        [model.inputs],
        [model.get_layer(last_conv_layer_name).output, model.output]
    )

    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        loss = predictions[:, 0]

    grads = tape.gradient(loss, conv_outputs)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

    conv_outputs = conv_outputs[0]
    heatmap = conv_outputs @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)

    return heatmap.numpy()

# ================================================================
# 17. Grad-CAM Visualization Example
# ================================================================

img_path = os.path.join(TEST_DIR, test_generator.filenames[0])
img = tf.keras.preprocessing.image.load_img(img_path, target_size=(IMG_SIZE, IMG_SIZE))
img_array = tf.keras.preprocessing.image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0) / 255.0

heatmap = make_gradcam_heatmap(img_array, transfer_model)

img = cv2.imread(img_path)
img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
heatmap_resized = cv2.resize(heatmap, (IMG_SIZE, IMG_SIZE))
heatmap_resized = np.uint8(255 * heatmap_resized)
heatmap_color = cv2.applyColorMap(heatmap_resized, cv2.COLORMAP_JET)

superimposed_img = cv2.addWeighted(img, 0.6, heatmap_color, 0.4, 0)

plt.figure(figsize=(6,6))
plt.imshow(cv2.cvtColor(superimposed_img, cv2.COLOR_BGR2RGB))
plt.title("Grad-CAM  Model Attention")
plt.axis('off')
plt.show()

# ================================================================
# 18. Save Transfer Learning Model
# ================================================================

transfer_model.save("pneumonia_detector_transfer_learning.h5")

print("Advanced model saved successfully!")

# ================================================================
# 19. Transfer Learning + Class Weights (Best Practice)
# ================================================================

from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import roc_curve, auc

# Compute class weights from training data
labels = train_generator.classes
class_weights_array = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(labels),
    y=labels
)

class_weights = {0: class_weights_array[0], 1: class_weights_array[1]}
print("Class Weights:", class_weights)

# Rebuild ResNet50 model (clean instance)
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3))

for layer in base_model.layers:
    layer.trainable = False

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(256, activation='relu')(x)
x = Dropout(0.5)(x)
output = Dense(1, activation='sigmoid')(x)

resnet_weighted_model = Model(inputs=base_model.input, outputs=output)

resnet_weighted_model.compile(
    optimizer=Adam(learning_rate=0.0001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

resnet_weighted_model.summary()

# Train with class weights

history_resnet_weighted = resnet_weighted_model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=8,
    class_weight=class_weights
)

# ================================================================
# 20. Evaluation + ROCAUC Curve
# ================================================================

pred_probs_resnet = resnet_weighted_model.predict(test_generator).ravel()

threshold = 0.7
pred_labels_resnet = (pred_probs_resnet > threshold).astype(int)
true_labels = test_generator.classes

print("Classification Report (ResNet50 + Class Weights):")
print(classification_report(true_labels, pred_labels_resnet, target_names=classes))

# ROC Curve
fpr, tpr, _ = roc_curve(true_labels, pred_probs_resnet)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(7,6))
plt.plot(fpr, tpr, label=f"ROC curve (AUC = {roc_auc:.3f})")
plt.plot([0,1], [0,1], linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROCAUC Curve (ResNet50)')
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# ================================================================
# 21. Grad-CAM for ResNet50 Model (Explainability)
# ================================================================

LAST_CONV_LAYER = "conv5_block3_out"

def make_gradcam_heatmap(img_array, model, last_conv_layer_name=LAST_CONV_LAYER):
    grad_model = tf.keras.models.Model(
        [model.inputs],
        [model.get_layer(last_conv_layer_name).output, model.output]
    )

    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        loss = predictions[:, 0]

    grads = tape.gradient(loss, conv_outputs)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

    conv_outputs = conv_outputs[0]
    heatmap = conv_outputs @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)

    return heatmap.numpy()

# Example Grad-CAM visualization
idx = 0
img_path = os.path.join(TEST_DIR, test_generator.filenames[idx])
img = tf.keras.preprocessing.image.load_img(img_path, target_size=(IMG_SIZE, IMG_SIZE))
img_array = tf.keras.preprocessing.image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0) / 255.0

heatmap = make_gradcam_heatmap(img_array, resnet_weighted_model)

orig_img = cv2.imread(img_path)
orig_img = cv2.resize(orig_img, (IMG_SIZE, IMG_SIZE))
heatmap_resized = cv2.resize(heatmap, (IMG_SIZE, IMG_SIZE))
heatmap_resized = np.uint8(255 * heatmap_resized)
heatmap_color = cv2.applyColorMap(heatmap_resized, cv2.COLORMAP_JET)

superimposed_img = cv2.addWeighted(orig_img, 0.6, heatmap_color, 0.4, 0)

plt.figure(figsize=(6,6))
plt.imshow(cv2.cvtColor(superimposed_img, cv2.COLOR_BGR2RGB))
plt.title("Grad-CAM  ResNet50 Attention Map")
plt.axis('off')
plt.show()

# ================================================================
# 22. Save Best Model
# ================================================================

resnet_weighted_model.save("pneumonia_detector_resnet50_weighted.h5")
print("Best model saved: pneumonia_detector_resnet50_weighted.h5")

```
